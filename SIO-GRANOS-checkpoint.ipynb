```python
# La cosecha no se encuentra armonizada, muchos agregan sólo el año de cosecha, mientras
# que otros escriben los dos años de la campaña. Por ejemplo, algunos escriben "COSECHA 2021" y otros "COSECHA 20/21",
# que es escencialmente lo mismo. Normalizo eso:

#Elimino los datos donde existe un error en la COSECHA, y se la ha cargado como "0":
data[data.COSECHA==0].count()
data = data[data.COSECHA!=0]

data.COSECHA = data.COSECHA.map(lambda x: x if x.find("/")>=0 else "COSECHA "+str(int(x[-2:])-1).zfill(2)+"/"+x[-2:].zfill(2))
data.COSECHA = data.COSECHA.map(lambda x: x.replace("-1", "99"))
data.COSECHA.unique()
```

```python
# Reemplazamos todos los "A Fijar" diferentes por un sólo tipo de "A Fijar":

data.loc[data.PRECIO.isin(['Fijar Precio/Prec. Cam.','Fijar Precio/Mercado Comprador',
'Fijar Precio/Otros', 'Fijar Precio/M. T. País', 'Fijar Precio','Fijar Precio/Fas teórico','Fijar Precio/PR R. Conj.',
 'Fijar Precio/M. T. Ext.']),"PRECIO"] = "A Fijar"


# Cambiamos la condición de pago a "A Plazo" y a "Contraentrega":
data.loc[~data["CONDICION PAGO"].isin(['A plazo']),"CONDICION PAGO"] = "Contraentrega"

# Cambiamos los tipos de contratos para armonizar las cancelaciones y rectificaciones:
data.loc[data.OPERACION.isin(['Anulación Fijación']),"OPERACION"] = "Anulación"
data.loc[data.OPERACION.isin(['Rectificación Fijación','Rectificación Ampliación']),"OPERACION"] = "Rectificación"

# Sacamos las anulaciones y rectificaciones, ya que no podemos darle seguimiento y las estaríamos duplicando.
# A las anulaciones podríamos tenerla en cuenta para períodos largos, como una campaña entera, pero para períodos cortos no,
# ya que una operación puede cancelarse meses después de ser dadas de alta.

data = data[~data.OPERACION.isin(["Anulación","Rectificación"])]

# Creo una dummy para ver si el comprador es final con destino a Córdoba, o si es una venta al exterior de Córdoba (final o no).
# Si la venta es a Córdoba pero NO es a un comprador final, no quiero tenerla en cuenta, porque 
# esa mercadería se volvería a comercializar y la contabilizaría dos veces. Si la mercadería sale de
# Córdoba, no me interesa si es final o no, porque no vuelve a aparecer en la base y ya salió de la provincia
# (ya que sólo tengo datos que provienen de Córdoba)
# DUMMY = "No usar", no me interesa. DUMMY = "Usar", la voy a usar en el análisis

cordoba = ["Cordoba/En destino","Zona 11/En destino","Zona 8/En destino","Zona 9/En destino",
          "Cordoba/En origen","Zona 11/En origen","Zona 8/En origen","Zona 9/En origen"]

# Indico que si no es final y tiene como destino a Córdoba me agregue que esta operacion no voy a usarla, de lo contrario sí
# voy a usarla:

data["DUMMY"] = np.where((data['ES FINAL']=="NO") & (data["LUGAR ENTREGA"].isin(cordoba)), 'No usar', 'Usar')

data


```

# Comercialización por producto y campaña

## Provincial

Cosas a tener en cuenta: Tengo que obtener el acumulado diario sin contar las retificaciones (porque no les puedo hacer seguimiento hacia atrás y el dato quedaría duplicado) ni los canjes. Tengo que descontar las anulaciones. Tengo que ver el acumulado comercializado, cuánto fue a precio hecho, cuánto a fijar, y de este último a cuánto ya se le puso precio.
Considerar que las operaciones hechas a fijar que todavía no le colocaron precio me van a distorcionar el precio promedio, por lo que tengo que tener un monto de toneladas a las que ya se les puso precio y otro de sin.

```python
# Elegir el período de análisis y el cultivo:

periodo_elegido = ""
cultivo_elegido = ""

periodo_elegido = input("Elegir la cosecha a analizar, con el formato 'COSECHA 20/21', etc.")
while periodo_elegido not in base_de_datos["COSECHA"].unique().tolist() and periodo_elegido!="exit":
    periodo_elegido = input("La fecha ingresada no se encontró, intentar nuevamente. Escribir 'exit' para salir.")

cultivo_elegido = input("Elegir el cultivo a analizar (TRIGO PAN, SOJA, MAIZ, GIRASOL, SORGO, ACEITE DE SOJA).")
while cultivo_elegido not in base_de_datos["PRODUCTO"].unique().tolist() and cultivo_elegido!="exit":
    cultivo_elegido = input("El cultivo ingresado no se encontró, intentar nuevamente o escribir 'exit' para salir.")


print(f"Se elegió el producto '{cultivo_elegido}' para la campaña '{periodo_elegido}'")
```

```python
# Filtrar el dataframe para dicha búsqueda:

base_filtrada = base_de_datos[base_de_datos["DUMMY"]=="Usar"]
base_filtrada = base_filtrada[base_filtrada["COSECHA"]==periodo_elegido]
base_filtrada = base_filtrada[base_filtrada["PRODUCTO"]==cultivo_elegido]

# Creamos la base de datos diaria TOTAL, con la agregación de todas las operaciones realizadas ese día:
base_filtrada_diaria = base_filtrada[["FECHA CONCERTACION","CANT. (TN)","MONTO TOTAL USD (MILES)"]].groupby("FECHA CONCERTACION").sum()

# Columna que marca las toneladas que se comercializaron con precio hecho:
base_filtrada_diaria['PRECIO HECHO']=base_filtrada[(base_filtrada["PRECIO"]=="Precio Hecho")&(base_filtrada["PRECIO/TN MONEDA"]!=0)][["FECHA CONCERTACION","CANT. (TN)"]].groupby("FECHA CONCERTACION").sum()["CANT. (TN)"]
base_filtrada_diaria['PRECIO HECHO'].fillna(0, inplace=True)

# Columna que marca las toneladas que se comercializaron con el precio a fijar en un futuro:
base_filtrada_diaria['A FIJAR']=base_filtrada[(base_filtrada["PRECIO"]=="A Fijar")][["FECHA CONCERTACION","CANT. (TN)"]].groupby("FECHA CONCERTACION").sum()["CANT. (TN)"]
base_filtrada_diaria['A FIJAR'].fillna(0, inplace=True)

# Columna que marca las toneladas que se comercializaron a fijar en algún momento y ese día se les colocó un precio:
base_filtrada_diaria['FIJADO']=base_filtrada[(base_filtrada["PRECIO"]=="A Fijar")&(base_filtrada["PRECIO/TN MONEDA"]!=0)][["FECHA CONCERTACION","CANT. (TN)"]].groupby("FECHA CONCERTACION").sum()["CANT. (TN)"]
base_filtrada_diaria['FIJADO'].fillna(0, inplace=True)

# Columna que acumula las cantidades comercializadas totales hasta ese fecha:
base_filtrada_diaria['ACUMULADO CANT. (TN)']=base_filtrada_diaria["CANT. (TN)"].cumsum(axis = 0)

base_filtrada_diaria.head()
```

```python
# Muestro los resultados sobre porcentaje de operaciones según modalidad.

comerc_precio_hecho = round(base_filtrada_diaria["PRECIO HECHO"].sum())
comerc_precio_a_fijar = round(base_filtrada_diaria["A FIJAR"].sum())
comerc_precio_fijado = round(base_filtrada_diaria["FIJADO"].sum())
comerc_precio_acumulado = round(max(base_filtrada_diaria["ACUMULADO CANT. (TN)"]))
fecha_filtrada_max = max(base_filtrada_diaria.index).strftime("%Y-%m-%d")


print(f"""
    A la fecha {fecha_filtrada_max} ya se comercializaron {comerc_precio_acumulado} toneladas 
    de {cultivo_elegido} para la campaña {periodo_elegido}. \n
    Del total, {str(round(comerc_precio_hecho*100/comerc_precio_acumulado,2))+"%"} se vendió con precio hecho
    y {str(round(comerc_precio_a_fijar*100/comerc_precio_acumulado,2))+"%"} con precio a fijar. \n
    De este último, a un {str(round(comerc_precio_fijado*100/comerc_precio_a_fijar,2))+"%"} ya se le colocó precio, 
    quedando {comerc_precio_a_fijar - comerc_precio_fijado} toneladas por colocar precio.
    """)

round(comerc_precio_hecho)
```
